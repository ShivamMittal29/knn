```python
#the most important thing in the knn is that  distance.so we can use the euclidian distance as well as the other distance as well
def distance(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
def knn(X,Y,querypoint,K=5):
    vals=[]#this array will store the distances
    m=X.shape[0]
    
    for i in range(m):
        d=distance(querypoint,X[i])
        vals.append((d,Y[i]))
    vals=sorted(vals)
    #nearest or thefirst k points
    vals=vals[:K]
    vals=np.array(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)
    index=new_vals[1].argmax()
    pred=new_vals[0][index]
        
    
    #print(vals)
    return pred
        
    
```

```python
x=knn(X,Y,query_x)
print(x)
```
